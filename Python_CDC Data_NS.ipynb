# Below I used groupby to see all topics that were captured

```python
df.groupby(['Topic'])['LocationDesc'].max()
```

# Based on the groupby, I grouped by topic and count of each per state, based on end year - this allowed me to see what topics each state encounters.

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
df.groupby(['Topic', 'LocationDesc'])['YearEnd'].sum()
```

```python
df.groupby(['Topic','YearEnd'])['LocationDesc'].count()
```

# Based on the topics, I then filtered out further to better understand how arthritis is evident in each state

```python
df1=df[df['Topic'] == 'Arthritis']
df1
```
